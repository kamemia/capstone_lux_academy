```python
import os
import caer
import canaro
import numpy as np
import cv2 as cv
import gc
```

```python
IMG_SIZE = (80, 80)
channels = 1
char_path = r'C:\Users\kamem\OneDrive\Desktop\Capstone_Lux_Academy\train'
```

```python
char_dict = {}
for char in os.listdir(char_path):
    char_dict[char] = len(os.listdir(os.path.join(char_path,char)))
```

```python
# Sort in descending order
char_dict = caer.sort_dict(char_dict, descending=True)
char_dict
```

```python
characters = []
count = 0
for i in char_dict:
    characters.append(i[0])
    count += 1
    if count>= 10:
        break
characters
```

```python
# Create the training data
train = caer.preprocess_from_dir(char_path, characters, channels=channels, IMG_SIZE=IMG_SIZE, isShuffle=True)
```

```python
# Number of training samples
len(train)
```

```python
# Visualizing the data (OpenCV doesn't display well in Jupyter notebooks)
import Image
import matplotlib.pyplot as plt
plt.figure(figsize=(30,30))
plt.imshow(train[0][0], cmap='gray')
plt.show()
```

```python
featureSet, labels = caer.sep_train(train, IMG_SIZE=IMG_SIZE)
```

```python
# Normalize the featureSet ==> (0, 1)
from tensorflow.keras.utils import to_categorical
featureSet = caer.normalize(featureSet)
labels = to_categorical(labels, len(characters))
```

```python
x_train, x_val, y_train, y_val = caer.train_val_split(featureSet, labels, val_ratio=.2)
```

```python
del train
del featureSet
del labels
gc.collect()
```

```python
BATCH_SIZE = 32
EPOCHS = 10
```

```python
# Image data generator
datagen = canaro.generators.imageDataGenerator()
train_gen = datagen.flow(x_train, y_train, batch_size=BATCH_SIZE)
```

```python
# Creating the model
model = canaro.createSimpsonsModel(IMG_SIZE=IMG_SIZE, channels=channels, output_dim=len(characters),
                                   loss='binary_crossentropy', decay=1e-6, learning_rate=0.001, momentum=0.9,
                                   nesterov=True)
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import LearningRateScheduler
callback_list = [LearningRateScheduler(canaro.lr_schedule)]
```

```python
training = model.fit(train_gen,
                     steps_per_epoch=len(x_train)//BATCH_SIZE,
                     epochs=EPOCHS,
                     validation_data=(x_val, y_val),
                     validation_steps=len(y_val)//BATCH_SIZE,
                     callbacks=callback_list)
```

```python
test_path = r''

img = cv.imread(test_path)

def prepare(img):
    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img = cv.resize(img, IMG_SIZE)
    img = caer.reshape(img, IMG_SIZE, 1)
    return img
```

```python
predictions = model.predict(prepare(img))
```

```python
# Getting class with the highest probability
print(characters[np.argmax(predictions[0])])
```
